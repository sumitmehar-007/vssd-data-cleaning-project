```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
#Loading our dataset using pandas library
vssd = pd.read_csv(r"F:\sumit\Vrinda Store Sales Dataset\Uncleaned Vrinda Store Sales Data.csv")               #vssd = Vrinda store sales dataset
print(vssd)
```

```python
#let's check how many row and coulmns in dataset
print("Row and Columns :",vssd.shape)
```

```python
#Let's chaeck info about dataset
vssd_info = vssd.info()                                                           #Here we see there in no any null values in any columns
```

```python
#let we count to confirm there is no null values
vssd_nv = vssd.isnull().sum()
print(vssd_nv)
#Here we see there is no need to fill any values because there is no null values in our dataset
```

```python
#Let check or Count duplicate values in dataset
vssd_duplicates = vssd.duplicated().sum()
print("Duplicate Values in Dataset : ",vssd_duplicates)
                                                                                        #Here we see there is no duplicate values in it
```

```python
#Let revise our data of top 10 persons and last 10 persons and Last 10 persons
vssd.head(10)
```

```python
vssd.tail(10)
```

```python
#We see as per our dataset ,in 'Qty' there is rendomly written '1' or 'One' , Lets replace it 'One' = '1' & correct 'Gender' column:
vssd["Qty"] = vssd["Qty"].replace("One","1")
vssd["Gender"] = vssd["Gender"].replace({"W" : "Female",
                                            "M" : "Male",
                                            "Men" : "Male",
                                             "Women": "Female"})
                                                                             #Let's check it replace or not?
print(vssd.head(10))
                                                                             
```

```python
#Let's check out data's profit or loss (mean,std,min,max...etc)
vssd.describe()
```

```python
vssd.rename(columns = {"Order ID" : "Order_Id",
              "Cust Id":"Cust_Id"},inplace = True)
vssd.head()
```

```python
# Here we observe we remove currency column because the currency in INR common in all orders
vssd.drop(["currency"],axis = 1)
```

```python
#Saving our CLeanned Data
vssd.head(10000).to_csv("Project1.csv",index=False)
print("File Saving Process Completed!!!")
```
